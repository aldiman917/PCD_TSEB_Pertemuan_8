```python
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
from skimage.util import invert

import numpy as np

#import library yang dibutuhkan
```

# Percobaan 1 - Cropping Image

```python
coffeeImage = data.coffee() #gambar dari library skimage. 
catImage = data.cat() #gambar dari library skimage. 

coffeeCropped = coffeeImage.copy() #Membuat salinan dari citra kopi 
coffeeCropped = coffeeCropped[0:256,64:320] #memotong citra mulai dari baris 0 hingga 255 (256 baris) dan dari kolom 64 hingga 319 (256 kolom). 

catCropped = catImage.copy() #Membuat salinan dari citra kucing
catCropped = catCropped[64:256,128:320] #emotong citra mulai dari baris 64 hingga 255 (192 baris) dan dari kolom 128 hingga 319 (192 kolom).

print('Coffee Ori Shape : ',coffeeImage.shape) #Menampilkan dimensi citra asli
print('Coffee Crop Shape : ',coffeeCropped.shape) #Menampilkan dimensi citra yang telah dipotong 

print('Cat Ori Shape : ',catImage.shape)
print('Cat Crop Shape : ',catCropped.shape)

fig, axes = plt.subplots(2, 2, figsize=(12, 12)) #Membuat plot menggunakan plt.subplots dengan grid 2x2 (4 subplot) dan ukuran figur 12x12.
ax = axes.ravel() #Mengubah array 2D axes menjadi 1D menggunakan ravel(). Hal ini mempermudah dalam mengakses setiap subplot.

ax[0].imshow(coffeeImage)
ax[0].set_title("Citra Input 1")

ax[1].imshow(catImage, cmap='gray')
ax[1].set_title('Citra Input 2')

ax[2].imshow(coffeeCropped)
ax[2].set_title("Citra Output 1")

ax[3].imshow(catCropped, cmap='gray')
ax[3].set_title('Citra Output 2')
```

# Percobaan 2 - Citra Negative

```python
inv = invert(coffeeCropped) #dilakukan proses invert pada citra coffeeCropped menggunakan fungsi invert
print('Shape Input : ', coffeeCropped.shape)
print('Shape Output : ',inv.shape) #citra hasil invertasi (inv)

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(coffeeCropped)
ax[0].set_title("Citra Input")

ax[1].hist(coffeeCropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(inv)
ax[2].set_title('Citra Output (Inverted Image)')

ax[3].hist(inv.ravel(), bins=256)
ax[3].set_title('Histogram Output') 
```

```python
copyCat = catCropped.copy().astype(float) #Membuat salinan dari citra catCropped
m1,n1 = copyCat.shape #Mengambil dimensi citra copyCat dan menyimpannya dalam variabel m1 dan n1.
output1 = np.empty([m1, n1]) #Membuat array kosong dengan dimensi yang sama seperti copyCat menggunakan np.empty() dan menyimpannya dalam variabel output1
for baris in range(0, m1-1): #Melakukan iterasi untuk setiap baris dalam citra.
    for kolom in range(0, n1-1): #Melakukan iterasi untuk setiap kolom dalam citra.
        a1 = baris #Menyimpan nilai baris dan kolom saat ini dalam variabel a1 dan b1.
        b1 = kolom
        output1[a1, b1] = copyCat[baris, kolom] + 100 #dilakukan proses peningkatan kecerahan pada citra catCropped dengan menambahkan nilai 100 pada setiap pikselnya.
        
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(catCropped, cmap='gray')
ax[0].set_title("Citra Input")

ax[1].hist(coffeeCropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(output1, cmap='gray')
ax[2].set_title('Citra Output (Brightnes)')

ax[3].hist(output1.ravel(), bins=256)
ax[3].set_title('Histogram Input')
```
