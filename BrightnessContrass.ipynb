# Import dependency

```python
#import librarry yang dibutuhkan
import numpy as np
import imageio
import matplotlib.pyplot as plt
import cv2
```

# Membaca gambar

```python
img = cv2.imread("tomat.jpg") #membaca gambar dengann file name tomat.jpg
```

# Mendapatkan resolusi dan type dari gambar

```python
#shape artinya mengakses dimensi dari img seperti (height, width, dan channel)
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
img_type = img.dtype #mengambil tipe data dari gambar yang sudah di-load dan menyimpannya ke dalam variabel img_type
```

# Brightness grayscale

```python
img_brightness = np.zeros(img.shape, dtype=np.uint8)
#np.zeros: NumPy dengar array baru kosong (zero)
#img.shape: shape dari array gambar, dimensi dari gambar (height, width, dan channel nya)
#dtype=np.uint8: spesifikasi type data array sebagai uint8
```

# Melakukan penambahan brightness dengan nilai yang menjadi parameter

```python


def brighter(nilai): #akan melakukan penambahan brightness dengan nilai yang akan ditentukan
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray += nilai
            if gray > 255: #batas atas nilai derajat keabuan
                gray = 255
            if gray < 0: #batas bawah niali derajat keabuan
                gray = 0
            img_brightness[y][x] = (gray, gray, gray) #karena mengubah ke grayscale maka (gray, gray, gray) semua parameternya

```

# Menampilkan beberapa hasil denan nilai brightness -100 dan 100

```python
brighter(-100) #brightness -100 artinya semakin redup (tanda negatif)
plt.imshow(img_brightness) #menampilkan gambar hasil brightness
plt.title("Brightness -100") #memberi tittle
plt.show()

brighter(100) #brightness 100 artinya semakin cerah (tanda positif)
plt.imshow(img_brightness) #menampilkan gambar hasil brightness
plt.title("Brightness 100") #memberi tittle
plt.show()
```

# Brightness RGB

# Membuat variabel img_rgbbrightness untuk menampung hasil

```python
img_rgbbrightness = np.zeros(img.shape, dtype=np.uint8)
#np.zeros: NumPy dengar array baru kosong (zero)
#img.shape: shape dari array gambar, dimensi dari gambar (height, width, dan channel nya)
#dtype=np.uint8:  mengatur tipe data dari array baru (img.rgbbrightness) sebagai uint8
```

# Melakukan penambahan brightness dengan nilai yg menjadi parameter

```python
def rgbbrighter(nilai): #penambahan intensitas warna yang ingin diterapkan pada gambar.
    for y in range(0, img_height):
        for x in range(0, img_width):
        red = img[y][x][0]
            red += nilai
            if red > 255: #nilai komponen warna merah tidak melebihi 255 
                red = 255                               
            if red < 0: #nilai komponen warna merah tidak kurang dari 0
                red = 0
            green = img[y][x][1] #mengambil nilai komponen warna hijau dari piksel yang sedang diproses.
            green += nilai
            if green > 255: #nilai komponen warna hijau tidak melebuhi dari 255
                green = 255
            if green < 0: #nilai komponen warna hijau tidak kurang dari 0
                green = 0
            blue = img[y][x][2] #mengambil nilai komponen warna biru dari piksel yang sedang diproses.
            blue += nilai
            if blue > 255: #nilai komponen warna biru tidak melebihi dari 255
                blue = 255
            if blue < 0: #nilai komponen warna biru tidak kurang dari 0
                blue = 0
            img_rgbbrightness[y][x] = (red, green, blue) #mengatur nilai komponen warna yang baru (setelah peningkatan intensitas) ke piksel yang sesuai dalam gambar img_rgbbrightness
```

# Menampilkan beberapa hasil dengan nilai brightness -100 dan 100

```python
rgbbrighter(-100) #-100, yang akan mengurangi intensitas warna sebesar 100 pada setiap komponen warna (red, green, blue) dari setiap piksel dalam gambar.
plt.imshow(img_rgbbrightness) #ditampilkan menggunakan plt.imshow dari library matplotlib
plt.title("Brightness -100") #tittle dari plot ditetapkan sebagai "Brightness -100".
plt.show() #menampilkan plot gambar yang telah dihasilkan.

rgbbrighter(100) #100. Ini akan menambahkan intensitas warna sebesar 100 pada setiap komponen warna dari setiap piksel dalam gambar 
plt.imshow(img_rgbbrightness)
plt.title("Brightness 100")
plt.show()
```


# Contrass
# 1. Membuat variabel img_contrass untuk menampung hasil

```python
img_contrass = np.zeros(img.shape, dtype=np.uint8)
#np.zeros: NumPy dengar array baru kosong (zero)
#img.shape: shape dari array gambar, dimensi dari gambar (height, width, dan channel nya)
#dtype=np.uint8:  mengatur tipe data dari array baru (img.contrass) sebagai uint8
```

# 2. Melakukan penambahan contrass dengan nilai yg menjadi parameter

```python
def contrass(nilai):
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray *= nilai
            if gray > 255: #nilai derajat keabuan tidak melebihi 255 (pada contrass tidak ada if gray>0, itu saja yang membedakan dengan brightness)
                gray = 255
            img_contrass[y][x] = (gray, gray, gray)
```

# Menampilkan beberapa hasil dengan nilai contrass 50 dan 100

```python
contrass(2) #nilai parameter 2, yang akan menggandakan kontras gambar. Ini berarti intensitas keabuan setiap piksel akan diperkuat dua kali lipat.
plt.imshow(img_contrass)
plt.title("Contrass 2")
plt.show()

contrass(3) #Fungsi contrass kembali dipanggil, kali ini dengan nilai parameter 3. Ini akan menggandakan kontras gambar lagi, menghasilkan intensitas keabuan yang lebih tinggi daripada sebelumnya.
plt.imshow(img_contrass)
plt.title("Contrass 3")
plt.show()
```


# Contrass Autolevel

# 1. Membuat variabel img_contrass untuk menampung hasil

```python
img_autocontrass = np.zeros(img.shape, dtype=np.uint8)
#np.zeros: NumPy dengar array baru kosong (zero)
#img.shape: shape dari array gambar, dimensi dari gambar (height, width, dan channel nya)
#dtype=np.uint8:  mengatur tipe data dari array baru (img_autocontrast) sebagai uint8
```

# 2. Melakukan penambahan contrass secara otomatis

```python
def autocontrass(): # bertujuan untuk secara otomatis meningkatkan kontras gambar.
    xmax = 300 #nilai keabuan maksimum
    xmin = 0 #nilai keabuan minimum
    d = 0
    # Mendapatkan nilai d, dimana nilai d ini akan berpengaruh pada hitungan
    # untuk mendapatkan tingkat kontras
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            if gray < xmax:
                xmax = gray
            if gray > xmin:
                xmin = gray
    d = xmin-xmax
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(float(255/d) * (gray-xmax))
            img_autocontrass[y][x] = (gray, gray, gray)
```

# 3. Menampilkan hasil autolevel contrass

```python
autocontrass()
plt.imshow(img_autocontrass)
plt.title("Contrass Autolevel")
plt.show()
```
